```sql
-- 1. Feladat
SELECT DATEDIFF(YEAR,SZUL_DAT,GETDATE()) AS 'Életkor', COUNT(USERNEV) AS 'Darab'
FROM Vendeg
WHERE DATEDIFF(YEAR,SZUL_DAT,GETDATE()) >= 18
GROUP BY DATEDIFF(YEAR,SZUL_DAT,GETDATE())
HAVING COUNT(USERNEV) >=5
```

```sql
-- 2. Feladat

SELECT SZOBA_FK AS 'Szoba azonosító', YEAR(METTOL) AS 'Év', COUNT(FOGLALAS_PK) AS 'Foglalások száma'
FROM Foglalas INNER JOIN Szoba ON Foglalas.SZOBA_FK = Szoba.SZOBA_ID
WHERE SZOBA_ID % 2 <> 0
GROUP BY SZOBA_FK, YEAR(METTOL)
HAVING COUNT(FOGLALAS_PK) > 10
```
