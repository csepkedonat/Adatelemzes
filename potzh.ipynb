```sql
SELECT (CASE GROUPING_ID(szallashely.HELY)
            when 1 then 'Összesen'
            when 0 then szallashely.HELY
            end) AS 'Hely'
, szallashely.CSILLAGOK_SZAMA, COUNT(FOGLALAS_PK)
FROM szallashely JOIN Szoba ON szallashely.SZALLAS_ID = szoba.SZALLAS_FK
JOIN Foglalas ON szoba.SZOBA_ID = Foglalas.SZOBA_FK
GROUP BY ROLLUP(Szallashely.HELY, szallashely.CSILLAGOK_SZAMA)
```

```sql
SELECT *, 
AVG(FELNOTT_SZAM+GYERMEK_SZAM) over(order by FOGLALAS_PK ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING) AS 'Átlag vendégszám'
FROM Foglalas
```
